```python
import pandas as pd
import numpy as np
import datetime as dt
```

```python
df1 = pd.read_csv('C:/Lamar_University/Spring_2023/Machine_Learning/Project_1/Land_Surface_Temperature/Data/statistics_LST_Day_1km_Crosby.csv')
df1.head()
```

```python
df1[['date','mean']]
```

```python
df1_R1 = df1[['date','mean']]
df1_R1['date'] = pd.to_datetime(df1_R1['date'])
df1_R1.set_index('date', inplace= True)
df1_R1.head(10)
```

```python
df1_R1_RS = df1_R1.resample('M').mean() #resampling Monthly
df1_R1_RS = df1_R1_RS.interpolate(method='linear')
df1_R1_RS
```

```python
df1_R1_RS.to_csv('Crosby_Temp.csv')
```
